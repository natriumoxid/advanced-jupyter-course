```python
import pandas as pd

# Display option
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

mydateparser = lambda x: pd.datetime.strptime(x, "%Y-%m-%d %H:%M:%S")
df = pd.read_csv('data/events.csv',  usecols=['quote_id', 'name_position', 'latitude', 'longitude', 'elevation', 'begin_timestamp', 'end_timestamp', 'codeset_description', 'average'], parse_dates=['begin_timestamp', 'end_timestamp'], date_parser=mydateparser)
df.head(10)
```

```python
df.dtypes
```

```python
filter = df[(df['elevation'] > 2250) & (df['elevation'] < 2300)]
filter.size
```
